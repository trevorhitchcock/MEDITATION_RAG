```python
import pandas as pd
import matplotlib.pyplot as plt

# Manually entered results
data = {
    "Model": ["Mistral-7B", "Llama-3-8B", "Flan-T5-Large"],
    "Semantic Similarity": [0.933, 0.933, 0.339],
    "Word Count": [475, 507, 59],
    "Sentence Count": [30, 37, 9],
    "TTR": [0.478, 0.393, 0.356]
}

df = pd.DataFrame(data)
```

```python
# Plot 1: Semantic Similarity
plt.figure()
plt.bar(df["Model"], df["Semantic Similarity"])
plt.title("Semantic Similarity by Model")
plt.ylabel("Cosine Similarity")
plt.ylim(0, 1)
plt.show()
```

```python
# Plot 2: Word Count
plt.figure()
plt.bar(df["Model"], df["Word Count"])
plt.title("Word Count by Model")
plt.ylabel("Number of Words")
plt.show()
```

```python
# Plot 3: Sentence Count
plt.figure()
plt.bar(df["Model"], df["Sentence Count"])
plt.title("Sentence Count by Model")
plt.ylabel("Number of Sentences")
plt.show()
```

```python
# Plot 4: Type-Token Ratio (TTR)
plt.figure()
plt.bar(df["Model"], df["TTR"])
plt.title("TTR by Model")
plt.ylabel("Lexical Diversity (TTR)")
plt.ylim(0, 1)
plt.show()
```
